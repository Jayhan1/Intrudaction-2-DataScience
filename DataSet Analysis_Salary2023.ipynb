```python
# -*- coding: utf-8 -*-
"""
Created on Sat Jun  3 10:07:24 2023

@author: Jihan Alnofli
"""
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV file
data = pd.read_csv(r'C:\Users\Hp\OneDrive\Desktop\ds_salaries.csv')

# Print information about the data
print("Number of rows:", data.shape[0])
print("Number of columns:", data.shape[1])
print("Sparsity:", data.isnull().sum().sum() / (data.shape[0] * data.shape[1]) * 100, "%")
print("Columns:", data.columns.tolist())
print("Data types:")
print(data.dtypes)
print(data)

# Visualize the data

# Bar chart for 'experience_level' column
data['experience_level'].value_counts().plot(kind='bar')
plt.title('Distribution of Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.show()

# Histogram for 'salary' column
data['salary'].plot(kind='hist', bins=10)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

# Pie chart for 'employment_type' column
data['employment_type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Employment Type Distribution')
plt.ylabel('')
plt.show()

# Scatter plot for 'salary' vs 'work_year'
plt.scatter(data['work_year'], data['salary'])
plt.title('Salary vs Work Year')
plt.xlabel('Work Year')
plt.ylabel('Salary')
plt.show()

# You can add more visualization charts based on your data and analysis needs



```
